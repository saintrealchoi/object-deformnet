```python
import numpy as np
import re

# Open the .obj file
with open('/data/obj_models/real_test/bottle_red_stanford_norm.obj', 'r') as f:
    lines = f.readlines()

# Extract the vertex coordinates from the .obj file
vertices = []
for line in lines:
    if line.startswith('v '):
        vertex = re.split('\s+', line.strip())[1:]
        vertex = [float(v) for v in vertex]
        vertices.append(vertex)
vertices = np.array(vertices)

# Compute the dimensions of the bounding box
min_point = np.min(vertices, axis=0)
max_point = np.max(vertices, axis=0)
dimensions = max_point - min_point

# Calculate the size of the bounding box
size = np.linalg.norm(dimensions)

# Print the size of the bounding box
print('Bounding box size:', size)
```

```python
import glob
import json

data_path = '/data/PhoCAL_release'
train_val = ['train','test']
train_model_set = set()
test_model_set = set()
# inst_list = glob.glob(os.path.join(obj_model_dir, subset, '*.obj'))
with open(os.path.join(data_path,'class_obj_taxonomy.json'),'r') as jpath:
    obj_file = json.load(jpath)
    
for split_data in train_val:
    for subset in os.listdir(os.path.join(data_path,split_data)):
        origin_path = os.path.join(data_path,split_data,subset,'origin','rgb_scene_gt.json')
        with open(origin_path,'r') as origin_path:
            sequence = json.load(origin_path)
        for scene_num in sequence:
            for idx,inst in enumerate(sequence[scene_num]):
                class_id = str(sequence[scene_num][idx]['class_id'])
                inst_id = str(sequence[scene_num][idx]['inst_id'])
                if split_data == 'train':
                    if obj_file[class_id]['objs'][inst_id] not in train_model_set:
                        train_model_set.add(obj_file[class_id]['objs'][inst_id])
                else:
                    if obj_file[class_id]['objs'][inst_id] not in test_model_set:
                        test_model_set.add(obj_file[class_id]['objs'][inst_id])
                    
        
        
        origin_path.close()
print(train_model_set)
print(test_model_set)
```
